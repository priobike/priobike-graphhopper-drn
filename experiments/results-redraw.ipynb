```python
import matplotlib.pyplot as plt

# This data was recorded by Max Lorenz in his Experiments for his Diploma Thesis.
# This script is just to re-plot these results.
plt_data = [
    # Error identifier, OSM, DRN
    ('a', 10, 0),
    ('b', 2, 0),
    ('c', 8, 0),
    ('d', 4, 1),
    ('e', 1, 0),
    ('f', 2, 0),
    ('g', 1, 0),
    ('h', 3, 1),
    ('i', 4, 1),
]

# Extract the data for the plot
labels = [x[0] for x in plt_data]
osm = [x[1] for x in plt_data]
drn = [x[2] for x in plt_data]

# Create a figure and an axis
fig, ax = plt.subplots(figsize=(6, 3))

# Set the width of the bars
width = 0.4

ax.grid(which="major", axis='x', color='#DAD8D7', alpha=0.5, zorder=0)
ax.grid(which="major", axis='y', color='#DAD8D7', alpha=0.5, zorder=0)

# Plot the OSM bars
ax.bar(labels, osm, width, color="#3498db", label='OSM', zorder=3)

# Plot the DRN bars next to the OSM bars
ax.bar([i + width for i in range(len(plt_data))], drn, width, color="#e74c3c",label='Our approach (Cycling Network Hamburg)', zorder=3)

# Add labels to the bars
for i, v in enumerate(osm):
    ax.text(i, v + 0.2, str(v), color='black', ha='center')

for i, v in enumerate(drn):
    ax.text(i + width, v + 0.2, str(v), color='black', ha='center')

# Add a legend and a title
ax.legend()
ax.set_title('Observed routing errors of test sample (N=72)')

ax.set_ylim(0, 12)

ax.set_ylabel('Counted number of errors')
ax.set_xlabel('Error identifier')

# Show the plot
plt.show()
# Save as pdf
fig.savefig('observed-errors.pdf', bbox_inches='tight')
```
